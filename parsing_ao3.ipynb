```python
base = 4901
i = base
while i < base+100:
    page = i
    real_html = requests.get('https://archiveofourown.org/tags/Harry%20Potter%20-%20J*d*%20K*d*%20Rowling/works?page={}'.format(page))
    print(page, real_html.status_code)
    if real_html.status_code != 200:
      break
    fname = "ao3_data_0"+str(int((base+100)/100))+".txt"
    with open (fname, 'a', encoding='utf-8') as f:
        json.dump(real_html.text, f, indent = 4, ensure_ascii = False)
    time.sleep(1)
    i+=1
```
